```python
import pandas as pd
import numpy as np
```

```python
movies=pd.read_csv('moviesds.csv')
```

```python
movies.head(10)
```

```python
movies.describe()
```

```python
movies.info()
```

```python
movies.isnull().sum()
```

```python
movies=movies[['id','title','overview','genres','keywords']]
movies
```

```python
movies.isnull().sum()
```

```python
movies.dropna(subset=['title'], inplace=True)
```

```python
movies.isnull().sum()
```

```python
movies['tag']=movies['genres']+movies['keywords']+movies['overview']
```

```python
movies=movies.drop(['overview','genres','keywords'],axis=1)
```

```python
movies['tag'] = movies['tag'].fillna('nothing')
```

```python
movies=movies[:20000]
movies.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=20000,stop_words='english')
```

```python
cv
```

```python
vector=cv.fit_transform(movies['tag'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
distance=sorted(list(enumerate(similarity[2])),reverse=True,key=lambda vector:vector[1])
```

```python
for i in distance[0:5]:
    print(movies.iloc[i[0]].title)
```

```python
def recommand(movie):
    index=movies[movies['title']==movie].index[0]
    distance=sorted(list(enumerate(similarity[index])),reverse=True,key=lambda vector:vector[1])
    for i in distance[0:15]:
        print(movies.iloc[i[0]].title)
```

```python
recommand("Thor")
```
